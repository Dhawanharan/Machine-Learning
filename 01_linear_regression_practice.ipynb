```python
import numpy as np
import matplotlib.pyplot as plt

#generate some fake data to practice
#X : hours studied, Y : exam marks
X = 2 * np.random.rand(100,1)
Y = 4 + 3 * X + np.random.randn(100,1)

plt.scatter(X,Y)
plt.xlabel("Hours Studied")
plt.ylabel("Marks")
plt.title("Our practice Data")
plt.show()
```

```python
#initialize our guesses
w = 0.0
b = 0.0

#set the learning rate
learning_rate = 0.01
epoches = 1000
n = len(X)

#the learning loop
for i in range(epoches):
  y_pred = w * X + b #predict y based on current  w and b

  dw = (2/n) * sum(X * (y_pred - Y))
  db = (2/n) * sum(y_pred - Y)

  w = w - learning_rate * dw
  b = b - learning_rate * db

  print(f"Learned Weight (m): {w[0]}")
  print(f"Learned Bias (c): {b[0]}")
```

```python
plt.scatter(X,Y)
plt.plot(X, w*X+b, color = 'red')
plt.show()
```
