```python
import tensorflow as tf
from tensorflow.keras import datasets, layers, models
import matplotlib.pyplot as plt

#1. Load the dataset
(train_images, train_labels), (test_images, test_labels) = datasets.cifar10.load_data()

#2. Normalize pixel values to be between 0 and 1
train_images, test_images = train_images / 255.0, test_images / 255.0

class_images = ['airplane', 'automobile', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']
```

```python
# building a convolutional neural network
model = models.Sequential([
    #layer 1:detect basic edges
    layers.Conv2D(32,(3,3), activation='relu', input_shape=(32,32,3)),
    layers.MaxPooling2D((2,2)),

    #layer 2:detects more complex shapes
    layers.Conv2D(64,(3,3), activation='relu'),
    layers.MaxPooling2D((2,2)),

    #layer 3: Final patterns
    layers.Conv2D(64,(3,3), activation='relu'),

    #Flattening to 1d for the final decision
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(10) # 10 outputs for 10 classes

])

model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

```python
#training
history = model.fit(train_images, train_labels, epochs=10, validation_data=(test_images, test_labels))
```
