```python
import tensorflow as tf
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing.sequence import pad_sequences

#1. load data
vocab_size = 10000
(X_train, y_train), (X_test, y_test) = imdb.load_data(num_words=vocab_size)

#2. padding
max_length = 250
X_train = pad_sequences(X_train, maxlen=max_length)
X_test = pad_sequences(X_test, maxlen = max_length)

print("Data loaded and padded!")
```

```python
#building an long short term memory
model = tf.keras.Sequential([
    #turns word into vectors
    tf.keras.layers.Embedding(vocab_size,32,input_length=max_length),

    #the memory layer
    tf.keras.layers.LSTM(64),

    #Final  Decision
    tf.keras.layers.Dense(1, activation='sigmoid') #gives score between 0 and 1
])
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

#TRain for epoches
model.fit(X_train, y_train, epochs=3, validation_data=(X_test, y_test), batch_size=64)
```
